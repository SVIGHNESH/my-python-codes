```python
import matplotlib.pyplot as plt

Student=[1,2,3,4,5]

MarksPy=[43,45,42,40,35]

MarksDs=[25,34,32,42,12]




x=range(len(Student))

bar_width=0.2

plt.bar([pos-bar_width/2 for pos in x], MarksPy, label='Python', width=bar_width)

plt.bar([pos+bar_width/2 for pos in x], MarksDs, label='Data Structure', width=bar_width)



plt.xlabel('Students')

plt.ylabel('Marks')

plt.title('Marks Comparison')

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


students = [1, 2, 3, 4, 5]
marks_python = [43, 45, 42, 40, 35]
marks_ds = [25, 34, 32, 42, 12]
marks_coa = [35, 23, 45, 10, 23]
marks_math = [38, 30, 45, 35, 28]
marks_english = [41, 38, 37, 33, 30]


bar_width = 0.15
x = np.arange(len(students))



plt.bar(x - 2*bar_width, marks_python, bar_width, label='Python', color='blue')
plt.bar(x - bar_width, marks_ds, bar_width, label='Data Structure', color='yellow')
plt.bar(x, marks_coa, bar_width, label='COA', color='pink')
plt.bar(x + bar_width, marks_math, bar_width, label='Mathematics', color='red')
plt.bar(x + 2*bar_width, marks_english, bar_width, label='English', color='black')


plt.xlabel('Students', fontsize=12)
plt.ylabel('Marks', fontsize=12)
plt.title('Student Marks Comparison Across Subjects', fontsize=14, pad=15)
plt.xticks(x, students)
plt.legend()

plt.tight_layout()

plt.show()
```

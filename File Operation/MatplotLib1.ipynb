```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(4, 500, 5)
y = np.tan(x)
plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt


x1 = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]


x2 = [1, 2, 3, 4, 5]
y2 = [3, 6, 9, 12, 15]


plt.plot(x1, y1, label='Line 1')


plt.plot(x2, y2, label='Line 2')


# plt.xlabel('X-axis')
# plt.ylabel('Y-axis')
plt.title('Two Line Graph')


plt.legend()


plt.show()
```
